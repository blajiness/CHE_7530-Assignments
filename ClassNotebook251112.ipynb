```python
Psdata = pd.read_csv('/content/MSE7530/sampledata/Beaucage_dataset.csv', header= None)
Psdata
```

```python
plt.scatter(Psdata[0], Psdata[1])
```

```python
theory_tg = 100
glass = Psdata[Psdata[0] < theory_tg]
liquid = Psdata[Psdata[0] >= theory_tg]
plt.scatter(glass[0], glass[1])
plt.scatter(liquid[0], liquid[1])
```

```python
mg, bg = np.polyfit(glass[0], glass[1], 1)
ml, bl = np.polyfit(liquid[0], liquid[1], 1)
```

```python
converge = (bl-bg)/(mg-ml)
converge
```

```python
x1 = np.linspace(60,84,24)
x2 = np.linspace(84,135,51)
```

```python
y1= mg * x1 + bg
y2= ml * x2 + bl
```

```python
theory_tg = 100
glass = Psdata[Psdata[0] < theory_tg]
liquid = Psdata[Psdata[0] >= theory_tg]
plt.scatter(glass[0], glass[1])
plt.scatter(liquid[0], liquid[1])
plt.plot(x1, y1, label='glass regression')
plt.plot(x2, y2, label='liquid regression')
plt.axvline(x=84,color='r', label='convergence X ~ 84')
plt.legend()
```
